```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Datasets/Iris.csv')
df.head()
```

```python
df.describe()
```

```python
grouped_df = df.groupby('Species').describe()
```

```python
grouped_df
```

```python
from scipy.stats import kurtosis
for column_name in df.columns[1:5]:
    grouped_kurtosis = df.groupby('Species')[column_name].apply(kurtosis)
grouped_kurtosis
```

```python
from scipy.stats import skew
for column_name in df.columns[1:5]:
    grouped_skewness = df.groupby('Species')[column_name].apply(skew)
grouped_skewness
```

```python
from scipy.stats import iqr
for column_name in df.columns[1:5]:
    grouped_iqr = df.groupby('Species')[column_name].apply(iqr)
grouped_iqr
```
